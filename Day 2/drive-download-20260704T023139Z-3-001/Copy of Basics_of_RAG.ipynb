<h1> Embeddings v1 <h1>


```python
import numpy as np

# Given sentences
sentences = ["the mouse ran up the clock", "the mouse ran down"]

# Step 1: Create the vocabulary
# Split sentences into words and flatten the list
words = [word for sentence in sentences for word in sentence.split()]
vocab = sorted(set(words))

print(vocab)
```

```python
# Step 2: Create word-to-index mapping
word_to_index = {word: i for i, word in enumerate(vocab)}

print(word_to_index)
```

```python
# Parameters for embeddings
vocab_size = len(vocab)  # Size of the vocabulary
embedding_dim = 64  # Dimensionality of the embeddings

# Step 3: Generate random embeddings
np.random.seed(42)  # For reproducibility
embeddings = np.random.rand(vocab_size, embedding_dim)
print(embeddings[:5, :4]) # just for print example purpose taking :5, :4
```

```python
# Step 4: Lookup function
def get_embedding(word):
    # Check if the word is in the vocabulary
    if word in word_to_index:
        word_idx = word_to_index[word]
        return embeddings[word_idx]
    else:
        return None
```

```python
# Example usage
word = "mouse"
embedding_vector = get_embedding(word)
print(f"Embedding for '{word}':\n{embedding_vector}")
```

<h1> Embeddings with Keras v2 <h1>


```python
# Using Keras


import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding

# Given sentences
sentences = ["the mouse ran up the clock", "the mouse ran down"]

# Step 1: Tokenize the text
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)
sequences = tokenizer.texts_to_sequences(sentences)

# Vocabulary size (adding 1 because of reserved 0 index)
vocab_size = len(tokenizer.word_index) + 1

# Step 2: Define the embedding layer within a model
embedding_dim = 64

model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=1, name="embedding_layer"),
])

# Dummy forward pass to initialize embeddings
# Note: This step is not necessary for embedding lookup but is shown here for completeness.
# Normally, you would train your model here.

# Step 3: Function to get embeddings
def get_embedding(word):
    try:
        # Convert the word to its index in the vocabulary
        word_index = tokenizer.word_index[word]
        # Get embeddings from the embedding layer
        embeddings = model.get_layer('embedding_layer').get_weights()[0]
        return embeddings[word_index]
    except KeyError:
        return None

# Example usage
word = "mouse"
embedding_vector = get_embedding(word)
print(f"Embedding for '{word}':\n{embedding_vector}")
```

<h1> Using Pre-trained Embeddings Models <h1>


```python
from sentence_transformers import SentenceTransformer

sentence = ["What you do, the way you think, makes you beautiful."]

# Initialize a SentenceTransformer model named 'paraphrase-MiniLM-L6-v2'
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

# Encode the list of sentences using the SentenceTransformer model
embedded_data = model.encode(sentence)

# Print the embedded data
print(embedded_data)
```
